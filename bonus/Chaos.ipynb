```python
from scipy.integrate import solve_ivp
```

```python
# INTRO - Basic population growth
Ncap = 10
L = 1 # lambda
def f(t,y):
  return L*(Ncap-y)

TMAX = 10
t = np.linspace(0,TMAX)
N0 = 2
ivp = solve_ivp(f, [0,TMAX],[N0], t_eval=t)

y = ivp.y[0]

plt.plot(t,y )
```

```python
# Figure 18.1 pg 270

def runmap(mu, x0=0.75, N=30):
  # Runs logistic map x_i+1 = f(x_i)
  x = np.zeros(N)
  x[0] = x0
  for i in range(0,N-1):
    x[i+1] = mu*x[i]*(1-x[i])

  return x

x0 = 0.9
N = 200
#N = 30
n = np.arange(N)
MULIST = [2.8, 3.3, 3.5, 3.8]
#MULIST = [3.827, 3.828, 3.829, 3.830]  # crit = 3.828427
#MULIST = [3.8282, 3.8283, 3.8284, 3.8285]  # crit = 3.828427
NUM = len(MULIST)

plt.figure(figsize=(16,4))
for i in range(NUM):
  mu = MULIST[i]
  x = runmap(mu,x0,N)
  plt.subplot(1,4,i+1)
  plt.plot(n,x)
  plt.title(f'$\mu$ = {mu}')
  plt.grid()
  plt.ylim(0,1)

plt.tight_layout()
plt.show()
```

```python
# BIFURCATION DIAGRAM: FIG 18.2, PAGE 273

MULIST = np.linspace(1,4, 1000)
MULIST = np.linspace(2.7,3.9, 1000)  # PART 8
MULIST = np.linspace(3.80,3.88, 1000)  # PART 10

MULIST = np.linspace(3.4,3.6, 1000)  # XXXXX

X0LIST = np.linspace(0.05,0.95, 50)
N = 200

xdata = []
mudata = []
for mu in MULIST:
  for x0 in X0LIST:
    x = x0
    for i in range(N):
      x = mu*x*(1-x)  # not storing all values
    mudata.append(mu)
    xdata.append(x)

plt.figure(figsize=(12,8))
plt.plot(mudata,xdata,',')
#plt.xlim(1,4)
plt.xlim(MULIST[0], MULIST[-1])
plt.ylim(0,1)
plt.xlabel('$\mu$')
plt.ylabel('$x_N$')
plt.title('BIFURCATION DIAGRAM')
plt.grid()
#plt.axvline(3.828427)  # part 10
plt.show()
```
